# Loading the book

```python
with open("miracle_in_the_andes.txt", "r", encoding = "utf-8") as file:
    book = file.read()
```

```python
import re
pattern = re.compile("[a-zA-Z]+")
findings = re.findall(pattern, book.lower())
dictionary = {}
for letter in findings:
    if letter in dictionary:
        dictionary[letter] += 1
    else: 
        dictionary[letter] = 1
tuple_sorted = sorted(dictionary.items(), key=lambda item: item[1], reverse = True)
dict_sorted = {}
for i, l in tuple_sorted:
    dict_sorted[i] = l
dict_sorted
```

```python
import nltk

from nltk.corpus import stopwords
english_stopwords = stopwords.words("english")
```

```python
english_stopwords
```

```python
for word in english_stopwords:
    if word in dict_sorted:
        del dict_sorted[word]
```

```python
list_dict = []
```

```python
for key, value in dict_sorted.items():
    list_dict.append((value, key))
```
